so it returns to the login page again , but what if i want to keep my session?

```python
with requests.session() as s :
    post_req=s.post(url=url_login,data=payload)
    print(post_req.url)
    r2=s.get(post_req.url)
    print(r2.url)
```

so we keep the same session.problem is solved.
